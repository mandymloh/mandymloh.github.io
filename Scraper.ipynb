```python
import requests
from bs4 import BeautifulSoup

url = "https://flavortotaste.com"
req = requests.get(url)
soup = BeautifulSoup(req.content, 'html.parser')
print(soup.prettify())
```

# Get recipe URLs

```python
for tag in soup.find_all('a', href=True):
    print(url + '/' + tag['href'].lstrip('./'))
```

# Get recipe titles

```python
text = soup.findAll('h5', text=True)

for i in range(len(text)):
    print(str(text[i]).lstrip('<h5 class="mb-0" style="padding-right: 10px; z-index: 1">').rstrip('</h5>'))
```

# Get relative image paths

```python
img_srcs = soup.findAll('img', src=True)
```

```python
github_raw_base_url = "https://raw.githubusercontent.com/mandymloh/mandymloh.github.io/master"

for i in range(len(img_srcs)):
    try:
        print('.' + str(img_srcs[i]).split(' src="./static')[1].split('" ')[0])
    except:
        pass
```
